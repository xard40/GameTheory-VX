```python
import pandas as pd
import matplotlib.pyplot as plt

# Get the season number from the user
try:
  season_number = int(input("Enter the season number: "))
except ValueError:
  print("Invalid season number. Please enter an integer.")
  exit()

# Get the 100 numbers as a string from the user
numbers_str = input("Enter 100 numbers separated by spaces: ")

# Convert the string to a list of numbers
try:
  numbers = [int(x) for x in numbers_str.split()]
  if len(numbers) != 100:
    raise ValueError("Please enter exactly 100 numbers.")
except ValueError as e:
  print(f"Invalid input: {e}")
  exit()

# Create a Pandas DataFrame
new_data = {'ID': numbers}
new_df = pd.DataFrame(new_data)

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(new_df['ID'], bins=30, edgecolor='black')

# Add labels and title
plt.title(f'Distribution of Season {season_number}')
plt.xlabel('ID')
plt.ylabel('Frequency')
plt.show()
```
